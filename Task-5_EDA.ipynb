```python
#write a condition to age

df = df[df["Age"] > 0]
```

```python
#check for outliers
#Box plot using purchase amount

plt.figure(figsize=(8,5))

sns.boxplot(x=df["Purchase Amount (USD)"])

plt.title("Purchase Amount Distribution")

plt.show()
```

```python
#check for outliers
#Box plot using age

plt.figure(figsize=(8,5))

sns.boxplot(x=df["Age"])

plt.title("Age Distribution")

plt.show()
```

```python
#univarient analysis for Age distribution

plt.figure(figsize=(8,5))

sns.histplot(df["Age"], bins=15)

plt.title("Age Distribution")

plt.xlabel("Age")

plt.ylabel("Count")

plt.show()
```

```python
#univarient analysis for Gender count

plt.figure(figsize=(6,4))

sns.countplot(x="Gender", data=df)

plt.title("Gender Distribution")

plt.show()
```

```python
#univarient analysis for Purchase amount distributio

plt.figure(figsize=(8,5))

sns.histplot(df["Purchase Amount (USD)"], bins=20)

plt.show()
```

```python
#univarient analysis for category count

plt.figure(figsize=(10,5))

sns.countplot(data=df,
              x="Category",
              order=df["Category"].value_counts().index)

plt.xticks(rotation=45)

plt.show()
```

```python
#univarient analysis for Review Rating

plt.figure(figsize=(8,5))

sns.histplot(df["Review Rating"], bins=10)

plt.show()
```

```python
#Bivarient analysis for gender V/s Purchase

plt.figure(figsize=(8,5))

sns.boxplot(x="Gender",
            y="Purchase Amount (USD)",
            data=df)

plt.show()
```

```python
#Bivarient analysis for Category V/s Purchase

plt.figure(figsize=(12,5))

sns.barplot(x="Category",
            y="Purchase Amount (USD)",
            data=df)

plt.xticks(rotation=45)

plt.show()
```

```python
#Bivarient analysis for Season V/s Purchase

plt.figure(figsize=(8,5))

sns.barplot(x="Season",
            y="Purchase Amount (USD)",
            data=df)

plt.show()
```

```python
#Bivarient analysis for Subscription Status V/s Purchase

plt.figure(figsize=(8,5))

sns.boxplot(x="Subscription Status",
            y="Purchase Amount (USD)",
            data=df)

plt.show()
```

```python
#Multivarient Analysis for category, gender & purchase amount

plt.figure(figsize=(12,6))

sns.barplot(data=df,
            x="Category",
            y="Purchase Amount (USD)",
            hue="Gender")

plt.xticks(rotation=45)

plt.show()
```

```python
#Scatter plot to check the relationship between purchase amount and age

plt.figure(figsize=(8,5))

sns.scatterplot(x="Age",
                y="Purchase Amount (USD)",
                hue="Gender",
                data=df)

plt.show()
```

```python
#Correlation for numerical data

numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(8,6))

sns.heatmap(numeric_df.corr(),
            annot=True,
            cmap="coolwarm")

plt.show()
```
